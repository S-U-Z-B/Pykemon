# API collection For RSQ3

RSQ3 - Main Question: Which individual Pokémon has the highest individual stat for each category? 

Sub questions: Which types of Pokémon have the highest base stat totals? Do abilities factor into a pokémons strength?

For this question we will need each base stat (hp, speed, attack etc.) in order to add them all together to create a base_stat_total result, and the name of the Pokemon. Within the API we get a lot of information per Pokemon, such as abilities and types which we will need for the sub questions; since there are 898 individual Pokemon this can be quite exhaustive.

## Task 1: Imports 

Below are the necessary imports for this notebook.

```python
import pandas as pd
import json
import requests #needed to call from the PokéAPI
import urllib.request
```

## Task 2: API Call

I call from the PokéAPI to get each individual pokemons url and get their information from there. I use a for loop to iterate through the 898 pokemons and get their urls, turn them into json file, and add them to the empty list 'json_responses'. Finally I turn the responses into a dataframe and print that information to show our full dataset.

PokéAPI site: https://pokeapi.co/

```python
#empty list to hold all json responses
json_responses = []
#loops through all 898 pokemon 
for x in range(898):
    #request each individual pokemons data (1-898)
    r = requests.get('http://pokeapi.co/api/v2/pokemon/{}/'.format(x+1))
    #adds all the json responses (dict) to the empty list
    json_responses.append(r.json())
    

#creates df from list which contains all responses
all_poke_df = pd.DataFrame(json_responses)
```
