# Data Prep For RSQ3: Base Stats + Total Base Stats

RSQ3 - Main Question: Which individual Pokémon has the highest individual stat for each category?

Sub questions: Which types of Pokémon have the highest base stat totals? Do abilities factor into a pokémons strength?

The aim for this notebook is to get the individual base stats give each one its own column. This will make it easier later on to answer the main question as I will have all the stat categories available in their own seperate columns and I will have the base stat total calculated too. 

## Task 1: Imports + Local File

Below are the necessary imports needed for this notebook.

```python
import pandas as pd
```

In the raw dataframe, the information needed for this question is within the 'stats' column. However this column contains a list of all 6 categories of stats per index which makes this more difficult to access.

```python
df_poke_raw = pd.read_csv("301_all_pokemon_data_edit.csv")
df_poke_raw
```

## Task 2: Create Base Stats Columns

Within the dataframe there is a 'stat' column which contains six seperate stat columns (HP, Attack, Defense, Special Attack, Special Defense, Speed). It is currently very difficult to get a given pokémons stats in this current format.

Below is an example of this format from the first pokémons stat column, we can see there are a lot of details present, however, the only information needed from this column is the base stat and the name of the category.

```python
df_poke_raw['stats'][0]
```

So, using the dataframe of every pokemon, I use a for loop to go through the 'stats' column and gather the six categories for easier analysis. I go into each category using a for loop and if statements. 

Since the categories are in order, I count the number of times we iterate through the categories and add the stat to its corresponding empty list. 

```python
#empty lists to hold eaach stat category
hp_base_stats = []
att_base_stats = []
def_base_stats = []
spAtt_base_stats = []
spDef_base_stats = []
speed_base_stats = []

for x in range(898):
    stat_res = df_poke_raw["stats"][x]
    #turns the dataframe column into a list, making it possible to loop through every name
    stat_res = eval(stat_res)
    i = 0
    #loops through a pokémons 6 stats
    for stat in stat_res:
        if i == 0:
            hp = stat["base_stat"]
            hp_base_stats.append(hp)
        if i == 1:
            attack = stat["base_stat"]
            att_base_stats.append(attack)
        if i == 2:
            defense = stat["base_stat"]
            def_base_stats.append(defense)
        if i == 3:
            spAttack = stat["base_stat"]
            spAtt_base_stats.append(spAttack)
        if i == 4:
            spDefense = stat["base_stat"]
            spDef_base_stats.append(spDefense)
        if i == 5:
            speed = stat["base_stat"]
            speed_base_stats.append(speed)
        i += 1
```

Next, I combine all six of the base stat lists into a zipped list which holds all their contained information.

```python
zippedList =  list(zip(hp_base_stats, att_base_stats, def_base_stats, spAtt_base_stats, spDef_base_stats, speed_base_stats))
```

Then, I turn this zipped list into a dataframe and include their title columns. 

```python
df_base_stats = pd.DataFrame(zippedList, columns = ['HP' , 'ATTACK', 'DEFENSE', 'SPECIAL_ATTACK', 'SPECIAL_DEFENSE', 'SPEED'])
df_base_stats
```

Now I will join the base stats data to the original dataframe. The new information is added onto the end of the dataframe.

```python
assert len(df_poke_raw) == len(df_base_stats)
```

```python
df_poke_stats = df_poke_raw.join(
    df_base_stats, 
    how='inner'
)
df_poke_stats
```

With these newly added columns on the dataframe, I decided to edit this further by removing several unnecessary columns such as 'held_items' and 'is_default', as well as the 'stats' column since I added the needed columns within it. 

```python
df_poke_stats = df_poke_stats.drop(columns = "base_experience")
df_poke_stats = df_poke_stats.drop(columns = "forms")
df_poke_stats = df_poke_stats.drop(columns = "game_indices")
df_poke_stats = df_poke_stats.drop(columns = "height")
df_poke_stats = df_poke_stats.drop(columns = "held_items")
df_poke_stats = df_poke_stats.drop(columns = "is_default")
df_poke_stats = df_poke_stats.drop(columns = "location_area_encounters")
df_poke_stats = df_poke_stats.drop(columns = "moves")
df_poke_stats = df_poke_stats.drop(columns = "order")
df_poke_stats = df_poke_stats.drop(columns = "weight")
df_poke_stats = df_poke_stats.drop(columns = "sprites")
df_poke_stats = df_poke_stats.drop(columns = "past_types")
df_poke_stats = df_poke_stats.drop(columns = "species")
df_poke_stats = df_poke_stats.drop(columns = "stats")
```

```python
df_poke_stats.columns
```

## Task 3: Calculate Base Stat Total

Next, I loop through all the pokémon in the dataframe and add all 6 stat categories together to get the base stat total. Then, I append each stat total to an empty list which I then turn into a dataframe and label as 'TOTAL_STAT'.

```python
total_base = []

for x in range(898):
    res = df_poke_stats['HP'][x] + df_poke_stats['ATTACK'][x] + df_poke_stats['DEFENSE'][x] + df_poke_stats['SPECIAL_ATTACK'][x] + df_poke_stats['SPECIAL_DEFENSE'][x] + df_poke_stats['SPEED'][x]   
    total_base.append(res)
```

```python
df_total = pd.DataFrame(total_base)
df_total.columns = ['TOTAL_STAT']
df_total
```

Now, I join the base stat total dataframe to the pokémon dataframe.

```python
df_poke_total = df_poke_stats.join(df_total, how= 'inner')
df_poke_total
```

## Task 4 : Save Local File 

Finally, I save this dataframe into a csv file which contains all the pokemons information as well as all the stats and its total.

```python
df_poke_total.to_csv("301_all_pokemon_data_base_stats_collect.csv", index=False)
df_poke_total.shape
```
