## Task 1: Imports and Local Files 

Below are the necessary imports needed for this notebook.

```python
import pandas as pd
```

Here I have the raw information that I called from the PokéAPI which contains the generations and the pokemons within that generation. I turn this local CSV file into a dataframe.

```python
df_gen_raw = pd.read_csv("401_all_generation_data.csv")
df_gen_raw
```

I also use the local file I created from RSQ3 which contains each individual pokemon and its relevant information, including stats. I turned this into a dataframe. My intention for this dataframe is to create a new column which will indicate the generation each pokemon is in by merging my findings.

```python
df_poke_raw = pd.read_csv("301_all_pokemon_data_base_stats_collect.csv")
df_poke_raw
```

## Task 2: Finding the Generations  

The relevant column I need to access is the 'pokemon_species' columns as it contains the name of all the pokémon in the generation. So, I loop through the 8 generations from the geenration dataframe and gather every name in said generation and place them into an empty list called 'names'. At the same time, I have a generation counter that I add to its own list called 'gen'. 

The lists for the pokemon and the generation number are then put into a zipped list so they can be displayed as a dataframe.

```python
#empty lists to collect 1) the pokemons name + 2) generation of the pokemon
names = []
gen = []
    
#loops through 8 generations
for x in range(8):
    genx = df_gen_raw["pokemon_species"][x]
    #turns the dataframe column into a list, making it possible to loop through every name
    genx = eval(genx)
    generation = x+1 #generation range of 1-8, to be appended to list
    #loops through pokemon names per generation
    for gx in genx:
        pokeName = gx["name"]
        names.append(pokeName)
        gen.append(generation)
```

```python
zippedList =  list(zip(names, gen))
```

```python
df_gen_no = pd.DataFrame(zippedList, columns = ['name', 'GENERATION_NO'])
df_gen_no
```

## Task 3: Merging Data

Now that we have the pokémons generation number, I want to combine the pokémon dataframe to the dataframe with the generation number. Both dataframes have a name column, thus we merge on this column and create a new dataframe which has all of the pokémons information with the addition of its generation number.

```python
df_pokeGen = df_poke_raw.merge(df_gen_no, on='name',how= 'left')
```

```python
df_pokeGen = df_pokeGen.set_index(df_pokeGen["id"])
```

```python
df_pokeGen
```

However, it appears that some of the pokémon from the pokémon dataframe were labelled incorrectly. As we can see, as I check the Generation number column for any null values, 26 pokemon come up without a generation. 

We can see from the 'name' column that all of these names were hyphenated by mistake, whereas in the generation number dataframe they were not hyphenated. As a result, there was no clear match and they ended up without a generation number.

    For example, the pokemon 'deoxys' (id 386) was named 'deoxys-normal'.

```python
df_pokeGen[df_pokeGen["GENERATION_NO"].isna()]
```

After looking through the 26 pokemon without a generation, I confirmed that it would be safe to fill in these null values by copying the previous value and entering it into the null index. Since the pokemon are in ID order which matches their generation order this makes this method easy to imploy.

I ensured that these pokemons were not the first pokemon in a given generation since this would cause an incorrect result.

```python
df_pokeGen = df_pokeGen.fillna(method ='pad')
```

To confirm that there are no longer any empty values I repeat the test to see if there are any null Generation numbers.

```python
df_pokeGen[df_pokeGen["GENERATION_NO"].isna()]
```

Finally, I enter one of the pokemons who were given a hyphenated name by mistake. In this instance, I checked on deoxys(-normal) and can confirm that their generation (Generation 3) was entered correctly.

```python
df_pokeGen[df_pokeGen["name"] == 'deoxys-normal']
```
